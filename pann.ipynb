```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = 'https://pann.nate.com/talk/348973119'
soup = BeautifulSoup(urlopen(url).read(), 'html.parser')
print(soup)
```

```python
print("=== 베스트 댓글 ===")
comment_best = soup.find("div", class_= "cmt_best")
print(comment_best.get_text())
#베스트 댓글들을 한꺼번에 가져온 다음에(리스트 한개로) 다시 3개로 나누어 줌
comment_best = comment_best.find_all('dd', class_ = "usertxt")
print(comment_best)
```

```python
for comment in comment_best:
    print("="*80)
    print(comment.get_text().strip())
print("="*80, "\n")
```

```python
comment_list = soup.find("div", class_ = "cmt_list")
print(comment_list.get_text())
comment_list = comment_list.find_all('dd', class_ = "usertxt")
print(comment_list)
```

```python
for comment in comment_list:
    print("="*80)
    print(comment.get_text().strip())
print("="*80, "\n")
```
